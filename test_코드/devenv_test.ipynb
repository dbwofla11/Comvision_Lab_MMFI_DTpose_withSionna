```python
import torch

print("==============================================")
print(f"🔥 PyTorch 버전: {torch.__version__}")
print(f"🚀 CUDA(GPU) 사용 가능 여부: {torch.cuda.is_available()}")

if torch.cuda.is_available():
    print(f"🛠️ PyTorch가 인식한 CUDA 버전: {torch.version.cuda}")
    print(f"💻 현재 연결된 짱짱한 GPU: {torch.cuda.get_device_name(0)}")
else:
    print("⚠️ 이런! PyTorch가 GPU를 인식하지 못하고 CPU로 도는 중입니다!")

print("----------------------------------------------")
# torchvision 설치 여부 및 버전 체크
try:
    import torchvision
    print(f"👁️ torchvision 버전: {torchvision.__version__} (✅ 잘 깔려있음!)")
except ImportError:
    print("❌ torchvision이 설치되어 있지 않습니다!")
print("==============================================")
```
