### 1번째 실험 Train Loss 시각화 (E01 - E03 baseline)


```python
import matplotlib.pyplot as plt

# =========================
# Figure 1 : Full Curve
# =========================

plt.figure(figsize=(12, 5))

plt.plot(
    train_epochs,
    train_losses,
    linewidth=2,
    label='Train Loss'
)

plt.plot(
    val_epochs,
    val_losses,
    marker='o',
    linewidth=2,
    label='Validation Loss'
)

plt.scatter(
    best_epoch,
    best_val,
    s=180,
    marker='*',
    zorder=5,
    label=f'Best Epoch {best_epoch}'
)

plt.title("Training & Validation Loss (Full)")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()


# =========================
# Figure 2 : Zoom-In
# =========================

plt.figure(figsize=(12, 5))

plt.plot(
    train_epochs,
    train_losses,
    linewidth=2,
    label='Train Loss'
)

plt.plot(
    val_epochs,
    val_losses,
    marker='o',
    linewidth=2,
    label='Validation Loss'
)

# Best point
plt.scatter(
    best_epoch,
    best_val,
    s=220,
    marker='*',
    zorder=10,
    label=f'Best Model ({best_epoch})'
)

# Annotation
plt.annotate(
    f'Best\nEpoch={best_epoch}\nLoss={best_val:.5f}',
    xy=(best_epoch, best_val),
    xytext=(145, 0.01732),
    arrowprops=dict(
        arrowstyle='->',
        lw=1.5
    ),
    fontsize=11
)

# 확대 범위
plt.xlim(0, 200)
plt.ylim(0.0167, 0.02)

plt.title("Training & Validation Loss (Zoom-In)")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
